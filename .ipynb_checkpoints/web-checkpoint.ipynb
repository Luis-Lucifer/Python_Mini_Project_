```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://www.amazon.com/s?k=playstation+5&sprefix=playstat%2Caps%2C392&ref=nb_sb_ss_ts-doa-p_1_8"
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36','Accept-Language': 'en-US,en;q=0.5'})
```

```python
webpage = requests.get(URL, headers=HEADERS)
```

```python
webpage
```

```python
soup = BeautifulSoup(webpage.content, 'html.parser')
```

```python
soup
```

```python
links = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
links[0]
```

```python
link = links[0].get('href')
link
```

```python
product_list = "https://amazon.com" + link
product_list
```

```python
new_webpage = requests.get(URL, headers=HEADERS)
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
new_soup
```

```python
hs=new_soup.find_all("span")
for i in range(0,len(hs)):
    print(str(hs[i])+'\n')
```

```python
title=[]
titlespan=new_soup.find_all("span",attrs={'class':'a-size-medium'})
for i in range(0,len(titlespan)):
    title.append(str(titlespan[i].text.strip()))
title[0]
len(title)
```

```python
prices=[]
pricespan=new_soup.find_all("span",attrs={'class':'a-price-whole'})
for i in range(0,len(pricespan)):
    prices.append(str(pricespan[i].text.strip().replace('.','$')))
prices[0]
len(prices)
```

```python
pricechart={}

for i in range(0,10):
    pricechart[title[i]]=prices[i]

pricechart
```

```python
if len(title)>len(prices):
    for i in range(0, len(title)-len(prices) ):
        title.pop()
if len(prices)>len(title):
    for i in range(0, len(prices)-len(title) ):
        prices.pop()

print(len(prices))
print(len(title))
```

```python
import pandas as pd

data = {'Product': title,
        'Price': prices} 
new = pd.DataFrame.from_dict(data)
new.to_csv('my_file.csv', index=False, header=True)
new
```
